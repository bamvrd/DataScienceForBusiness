#One-hot-encoding : Chiang Rai's details column

```python
merge_ch2 = pd.read_csv('/content/merge_ch_nofill.csv')
#merge_nk2 = pd.read_csv('/content/merge_nk3.csv')
```

```python
merge_ch2.columns
```

```python
#split string by comma and slash 
merge_ch2['details'] =  merge_ch2['details'].str.strip('[]').str.split("/")
#one-hot encoding
ch_ohe = (
    merge_ch2['details'].explode()
    .str.get_dummies().sum(level=0).add_prefix('detail_')
)
merge_ch2 = merge_ch2.drop('details', 1).join(ch_ohe)
```

#One-hot-encoding : Chiang Rai's cnd_flr_no

```python
for i in range(len(merge_ch2['cnd_flr_no'])):
  if "," in str(merge_ch2['cnd_flr_no'][i]):
    # print(merge_ch2['cnd_flr_no'][i])
    merge_ch2['cnd_flr_no'][i] = merge_ch2['cnd_flr_no'][i].replace(", ", ",")

merge_ch2['cnd_flr_no'] =  merge_ch2['cnd_flr_no'].astype(str).str.strip('[]').str.split("/")

ch_ohe = (
    merge_ch2['cnd_flr_no'].explode()
    .str.get_dummies().sum(level=0).add_prefix('flr_')
)
merge_ch2 = merge_ch2.drop('cnd_flr_no', 1).join(ch_ohe)

merge_ch2
```

#One-hot-encoding : Nakhon's details column

```python
for i in range(len(merge_nk2['details'])):
  merge_nk2['details'][i] = merge_nk2['details'][i].replace("พื้นทีระเบียง", "พื้นที่ระเบียง").replace(" และ", "/").replace(" พื้นที่", "/พื้นที่").replace("และ", "/").replace(",", "/")

merge_nk2['details'] =  merge_nk2['details'].str.strip('[]').str.split("/")

nk_ohe = (
    merge_nk2['details'].explode()
    .str.get_dummies().sum(level=0).add_prefix('detail_')
)
merge_nk2 = merge_nk2.drop('details', 1).join(nk_ohe)
```

#One-hot-encoding : Nakhon's cnd_flr_no column

```python
for i in range(len(merge_nk2['cnd_flr_no'])):
  if "," in str(merge_nk2['cnd_flr_no'][i]):
    merge_nk2['cnd_flr_no'][i] = merge_nk2['cnd_flr_no'][i].replace(", ", ",")

merge_nk2['cnd_flr_no'] =  merge_nk2['cnd_flr_no'].astype(str).str.strip('[]').str.split(",")

nk_ohe = (
    merge_nk2['cnd_flr_no'].explode()
    .str.get_dummies().sum(level=0).add_prefix('flr_')
)
merge_nk2 = merge_nk2.drop('cnd_flr_no', 1).join(nk_ohe)

merge_nk2
```

#add another feature : distance to populated place
## populated place in Chiang Rai = Central Chiang Rai, Clock Tower
## populated place in Nakhon = Central, terminal, Khao Yai


```python
from math import radians, cos, sin, asin, sqrt
def distance(lat1, lat2, lon1, lon2):
    # The math module contains a function named
    # radians which converts from degrees to radians.
    lon1 = radians(float(lon1))
    lon2 = radians(float(lon2))
    lat1 = radians(float(lat1))
    lat2 = radians(float(lat2))
      
    # Haversine formula
    dlon = lon2 - lon1
    dlat = lat2 - lat1
    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
 
    c = 2 * asin(sqrt(a))
    
    # Radius of earth in kilometers. Use 3956 for miles
    r = 6371
      
    # calculate the result
    return(c * r)
```

```python
# add 3 new rows : 'distance_from_condo_to_terminal', 
#                  'distance_from_condo_to_central', 
#                  'distance_from_condo_to_khao_yai'
nk_terminal_lat = 14.98157186688729
nk_terminal_lng = 102.09035432330184
nk_central_lat = 14.995922395816116
nk_central_lng = 102.11678196196969
nk_khao_yai_lat = 14.42670535694956
nk_khao_yai_lng = 101.37887350844382
merge_nk2['distance_from_condo_to_terminal'] = 0.0
merge_nk2['distance_from_condo_to_central'] = 0.0
merge_nk2['distance_from_condo_to_khao_yai'] = 0.0
for row in range(len(merge_nk2)):
    merge_nk2['distance_from_condo_to_terminal'][row] = distance(nk_terminal_lat, merge_nk2['lat'][row], nk_terminal_lng, merge_nk2['lng'][row])
    merge_nk2['distance_from_condo_to_central'][row] = distance(nk_central_lat, merge_nk2['lat'][row], nk_central_lng, merge_nk2['lng'][row])
    merge_nk2['distance_from_condo_to_khao_yai'][row] = distance(nk_khao_yai_lat, merge_nk2['lat'][row], nk_khao_yai_lng, merge_nk2['lng'][row])
```

```python
# add 2 new rows : 'distance_from_condo_to_central',                   
#                  'distance_from_condo_to_clock'
ch_central_lat = 19.88668352168141
ch_central_lng = 99.83472886662774
ch_clock_lat = 19.907142095365796
ch_clock_lng = 99.83096894481031
merge_ch2['distance_from_condo_to_central'] = 0.0
merge_ch2['distance_from_condo_to_clock'] = 0.0
for row in range(len(merge_ch2)):
    merge_ch2['distance_from_condo_to_central'][row] = distance(ch_central_lat, merge_ch2['lat'][row], ch_central_lng, merge_ch2['lng'][row])
    merge_ch2['distance_from_condo_to_clock'][row] = distance(ch_clock_lat, merge_ch2['lat'][row], ch_clock_lng, merge_ch2['lng'][row])
```
